```
import pandas as pd
```

```
year = pd.read_excel('doc.xlsx', parse_dates=['Date'])
year 
```

```
series = pd.to_datetime(year['Date'], format='%d/%m/%Y')
series
```

```
year['Day of week (int)'] = year['Date'].dt.weekday

year['Day of week (str)'] = year['Date'].dt.day_name()

print(year)
```

```
year = year.loc[year["Day of week (int)"] == 0]
```

```
year.to_excel('mondays.xlsx')
```

```
year.to_csv('mondays.csv')
```
